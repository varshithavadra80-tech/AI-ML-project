```python

import ollama

model = "llama3:8b"

def ask_llm(inp):

    rules = """
    You are a professional Financial Assistant.

    Responsibilities:
    - Help users create budgets and savings plans.
    - Explain financial concepts in simple language.
    - Provide investment education.
    - Assist with debt repayment strategies.
    - Help calculate interest, EMI, returns, and expenses.
    - Suggest ways to reduce spending and improve savings habits.

    Rules:
    - Never claim to be a licensed financial advisor.
    - Do not guarantee profits or returns.
    - Mention risks associated with investments.
    - Avoid legal and tax advice.
    - Keep responses simple and practical.
    """

    response = ollama.chat(
        model=model,
        messages=[
            {
                "role": "system",
                "content": rules
            },
            {
                "role": "user",
                "content": inp
            }
        ],
        stream=True
    )

    for res in response:
        print(res["message"]["content"], end="", flush=True)

    print("\n")


while True:
    inp = input("Enter your financial question: ")

    if inp.lower() in ["exit", "quit"]:
        print("Goodbye!")
        break

    ask_llm(inp)    
```

```python
from langchain_community.document_loaders import PyPDFLoader

FILE_PATH = r"C:\Users\admin\Downloads\varshitha.pdf"
S
loader = PyPDFLoader(FILE_PATH)
docs = loader.load()

print(docs[0].page_content)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=60
)
texts = text_splitter.split_documents(docs)
```
